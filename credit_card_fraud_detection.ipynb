```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python


df=pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df["Class"].value_counts()
```

```python
legit_transaction=df[df["Class"]==0]
```

```python
fraud_transaction=df[df["Class"]==1]
```

```python
fraud_transaction.shape
```

```python
legit_transaction_sample=legit_transaction.sample(500)
```

```python
legit_transaction_sample.shape
```

```python
new_df=pd.concat([legit_transaction_sample,fraud_transaction])
new_df.shape
```

```python
new_df["Class"].value_counts()
```

```python
X=new_df.iloc[:,:-1]
X.shape
```

```python
y=new_df["Class"]
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression(max_iter=100)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
